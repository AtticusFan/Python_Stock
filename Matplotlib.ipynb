```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
pic = np.linspace(-10,10,100)
#定義一個圖像窗口
fig = plt.figure()
#定義x,y和圖的樣式
plt.plot(pic, np.sin(pic), '.') 
fig2 = plt.figure()
plt.plot(pic, np.sin(pic), '.', pic, np.cos(pic), '--')
```

```python
pic_error_x = np.linspace(-10,10,50)
pic_error_y = np.random.randint(-10,10,50)
fig = plt.figure()
plt.errorbar(pic_error_x, pic_error_y, yerr=2, fmt='o', color='SteelBlue', ecolor='LightSteelBlue', elinewidth=2)
# ecolor->誤差的顏色  elinewidth->誤差的寬度
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('classic')
%matplotlib inline
graph = np.linspace(0,10, 1000)  #產生 0~10區間的 等間隔序列
fig , ax = plt.subplots()
plt.plot(graph, np.sin(graph), '-', color='Blue', label='Sine')
plt.plot(graph, np.cos(graph), '-', color='Red', label='Cosine')
ax.axis('equal')
# loc(名稱放在圖表的右上角) frameon(名稱是否用框框) shadow(是否有陰影)
leg = ax.legend(loc='upper right', shadow=False) 
```
